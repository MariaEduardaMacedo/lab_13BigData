```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col , upper , avg

spark = SparkSession.builder.appName("ExerciciosSpark").getOrCreate()

df = spark.read.csv("DataFrame.csv",header=True,inferSchema=True)
print("5 primeiras linhas do DataFrame")
df.show(5)

print("Pessoas com mais de 25 anos")
df.filter(col("Idade") > 25).show()

print("Selecionando apenas Nome e Cidade")
df_selecionar=df.select("Nome","Cidade")
df_selecionar.show()

df_ordenar=df.orderBy(col("Idade").desc())
print("Ordem decrescente de Idade")
df_ordenar.show()

df_maiusculo=df.withColumn("Nome",upper(col("Nome")))
print("Coluna Nome em maiúsculo")
df_maiusculo.show()

idade_media=df.agg(avg("Idade").alias("Idade Média"))
print("Idade média")
idade_media.show()

pessoas_por_cidade=df.groupBy("Cidade").count().alias("Contagem")
print("Contagem de pessoas por cidade")
pessoas_por_cidade.show()

```
